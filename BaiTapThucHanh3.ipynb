# RNN đơn giản (one-to-one, many-to-one)

```python
import tensorflow as tf

# Cấu hình mô hình RNN nhiều lớp (many-to-one)
num_classes = 5  # ví dụ 5 lớp phân loại
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=5000, output_dim=32),   # Embedding từ từ thành vector
    tf.keras.layers.SimpleRNN(64),                              # RNN layer với 64 đơn vị ẩn
    tf.keras.layers.Dense(num_classes, activation='softmax')    # Lớp output với softmax
])
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
import re
import unicodedata
from underthesea import word_tokenize

def normalize_unicode(text):
    """Chuẩn hóa mã hóa Unicode -> NFC"""
    return unicodedata.normalize("NFC", text)

def remove_urls(text):
    return re.sub(r'http\S+|www\S+|https\S+', '', text)

def remove_emojis(text):
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF"
        u"\U0001F1E0-\U0001F1FF"
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)

def remove_punctuation(text):
    return re.sub(r"[^0-9a-zA-ZÀ-ỹà-ỹ\s]", " ", text)
def clean_spaces(text):
    return re.sub(r"\s+", " ", text).strip()
def preprocess(text):
    text = normalize_unicode(text)
    text = text.lower()
    text = remove_urls(text)
    text = remove_emojis(text)
    text = remove_punctuation(text)
    text = clean_spaces(text)
    text = word_tokenize(text, format="text")
    return text
```

```python
sample = "Giáo viên RẤT nhiệt tình!!! Tôi học Data!!!"

print("Before:")
print(sample)
print("\nAfter:")
print(preprocess(sample))
```

```python
def preprocess_dataset(data, key='sentence'):
    cleaned = []
    for item in data:
        text = preprocess(item[key])
        cleaned.append(text)
    return cleaned

processed_train = preprocess_dataset(train)
processed_train[:10]
```

### Bài 1: Xây dựng mạng LSTM gồm 5 lớp với hidden size là 256 cho bài toán phân loại văn bản. Huấn luyện mô hình này trên bộ dữ liệu UIT-VSFC (Vietnamese Student Feedback Corpus) sử dụng Adam làm phương thức tối ưu tham số và đánh giá độ hiệu quả của mô hình sử dụng độ đo F1.

```python
label_map = {
    "negative": 0,
    "neutral": 1,
    "positive": 2
}

def prepare_xy(dataset):
    texts = [preprocess(sample["sentence"]) for sample in dataset]
    labels = []

    for sample in dataset:
        lbl = sample["sentiment"]
        labels.append(label_map[lbl])

    return texts, np.array(labels, dtype="int32")


X_train_texts, y_train = prepare_xy(train)
X_dev_texts,   y_dev   = prepare_xy(dev)
X_test_texts,  y_test  = prepare_xy(test)

print(X_train_texts[0])
print("Label:", y_train[0])
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
MAX_WORDS = 20000
MAX_LEN   = 60

tokenizer = Tokenizer(num_words=MAX_WORDS, oov_token="<OOV>")
tokenizer.fit_on_texts(X_train_texts)

def texts_to_padded(texts):
    seqs = tokenizer.texts_to_sequences(texts)
    padded = pad_sequences(seqs, maxlen=MAX_LEN, padding='post', truncating='post')
    return padded

X_train = texts_to_padded(X_train_texts)
X_dev   = texts_to_padded(X_dev_texts)
X_test  = texts_to_padded(X_test_texts)

vocab_size = min(MAX_WORDS, len(tokenizer.word_index) + 1)
num_classes = len(np.unique(y_train))

print("Vocab size:", vocab_size)
print("Num classes:", num_classes)
print("X_train shape:", X_train.shape)
```

```python
import tensorflow as tf

HIDDEN = 256

model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, HIDDEN_SIZE, input_length=MAX_LEN),

    tf.keras.layers.LSTM(HIDDEN_SIZE, return_sequences=True),
    tf.keras.layers.LSTM(HIDDEN_SIZE, return_sequences=True),
    tf.keras.layers.LSTM(HIDDEN_SIZE, return_sequences=True),
    tf.keras.layers.LSTM(HIDDEN_SIZE, return_sequences=True),

    tf.keras.layers.LSTM(HIDDEN_SIZE),   # cuối cùng
    tf.keras.layers.Dense(num_classes, activation='softmax')
])
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

```

```python
BATCH_SIZE = 64
EPOCHS = 5

history = model.fit(
    X_train, y_train,
    validation_data=(X_dev, y_dev),
    batch_size=BATCH_SIZE,
    epochs=EPOCHS
)
```

```python
from sklearn.metrics import f1_score

# Dev
y_dev_pred_prob = model.predict(X_dev)
y_dev_pred = np.argmax(y_dev_pred_prob, axis=1)
f1_dev = f1_score(y_dev, y_dev_pred, average='macro')

# Test
y_test_pred_prob = model.predict(X_test)
y_test_pred = np.argmax(y_test_pred_prob, axis=1)
f1_test = f1_score(y_test, y_test_pred, average='macro')

print("F1-macro (Dev): ", f1_dev)
print("F1-macro (Test):", f1_test)
```

```python
from collections import Counter

word_counter = Counter()
tag_counter = Counter()

for sent in train:
    for w in sent["words"]:
        word_counter[w] += 1
    for t in sent["tags"]:
        tag_counter[t] += 1
```

```python
word2id = {"<PAD>":0, "<UNK>":1}
for w,_ in word_counter.most_common():
    word2id[w] = len(word2id)

id2word = {v:k for k,v in word2id.items()}
```

```python
tag2id = {"<PAD>":0}
for t,_ in tag_counter.most_common():
    tag2id[t] = len(tag2id)

id2tag = {v:k for k,v in tag2id.items()}
```

```python
import numpy as np

MAX_LEN = 50

def encode_sent(words, tags):
    # token → id
    ids = [word2id.get(w,1) for w in words]
    tg  = [tag2id[t] for t in tags]

    # pad
    if len(ids)<MAX_LEN:
        pad_len = MAX_LEN - len(ids)
        ids += [0]*pad_len
        tg  += [0]*pad_len
    else:
        ids = ids[:MAX_LEN]
        tg  = tg[:MAX_LEN]

    return ids, tg

def build_dataset(data):
    X, Y = [], []
    for sample in data:
        a,b = encode_sent(sample["words"], sample["tags"])
        X.append(a); Y.append(b)
    return np.array(X), np.array(Y)

X_train, y_train = build_dataset(train)
X_dev,   y_dev   = build_dataset(dev)
X_test,  y_test  = build_dataset(test)

print(X_train.shape, y_train.shape)
print(len(word2id), len(tag2id))
```

```python
import tensorflow as tf

HIDDEN = 256
VOCAB_SIZE = len(word2id)
TAG_SIZE = len(tag2id)

model = tf.keras.Sequential([
    tf.keras.layers.Embedding(VOCAB_SIZE, 256, input_length=MAX_LEN),

    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(HIDDEN, return_sequences=True)),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(HIDDEN, return_sequences=True)),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(HIDDEN, return_sequences=True)),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(HIDDEN, return_sequences=True)),

    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(HIDDEN, return_sequences=True)),

    tf.keras.layers.Dense(TAG_SIZE, activation="softmax")
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

```python
BATCH = 64
EPOCH = 4

history = model.fit(
    X_train, y_train,
    validation_data=(X_dev, y_dev),
    batch_size=BATCH,
    epochs=EPOCH
)
```
